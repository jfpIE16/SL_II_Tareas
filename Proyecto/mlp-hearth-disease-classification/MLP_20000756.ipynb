A continuación se muestra la información y significado de cada una de las características del dataset:  
1. age, edad
2. sex, género
3. cp, tipo de dolor en el pecho
    * 1: typical angina
    * 2: atypical angina
    * 3: non-anginal pain
    * 4: asymptomatic
4. trestbps, presión sanguínea en reposo (mm Hg)
5. chol, colesterol serico (mg/dl)
6. fbs, glucemia en ayunas (>120 mg/dl)
7. restecg, resultados electrocardiográficos en reposo
    * 0: normal
    * 1: anomalía en la onda ST-T
    * 2: hipertrofía ventricular izquierda probable o definitiva según los criterios de Estes'
8. thalach, frecuencia cardíaca máxima alcanzada
9. exang, angina inducida por ejercicio (1=si, 0=no)
10. oldpeak, depresión del ST inducida por el ejercicio en relación con la prueba
11. slope, la pendiente del segmento ST de ejercicio pico
    * 1: subiendo
    * 2: plana
    * 3: descendente
12. ca, número de vasos mayores (0-3)
13. thal,3 = normal; 6 = defecto fijo; 7 = defecto reparable
14. num(target), diagnostico de enfermedad del corazón (estado de la enfermedad angiográfica)
    * 0: Negativo
    * 1: Positivo

**Preparamos los datos para entrenamiento**

```python
X = hearth_data.drop('target', axis=1)
Y = hearth_data['target']

#Dimensiones del conjunto de datos
print(X.shape)
print(Y.shape)
```

```python
X_train, X_test, y_train,  y_test = train_test_split(X, Y, test_size=0.3, shuffle=True, random_state=756)

print("Dimensiones datos de entrenamiento:")
print (X_train.shape, y_train.shape)
print("Dimensiones datos de prueba:")
print (X_train.shape, y_test.shape)
```

**Nota:** Se procederá a normalizar el conjunto de datos y observaremos su impacto en el modelo final.

```python
scaler = MinMaxScaler()
X_train_norm = scaler.fit_transform(X_train)
X_test_norm = scaler.fit_transform(X_test)

X_train = pd.DataFrame(X_train_norm)
X_test = pd.DataFrame(X_test_norm)
```

### Entrenando una red neuronal utilizando Keras

```python
#Definimos callbacks de utilidad
#Learning rate decay
reducelrp = ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=30, verbose=0, mode='auto', min_delta=0.0001, cooldown=0, min_lr=0)
#Early stopping
earlystop_monitor = EarlyStopping(monitor='val_loss', verbose=1, patience=30, baseline=0.4,)
#Model checkpoint
best_model = 'mlp_hearthdisease.h5'
checkpoint = ModelCheckpoint(best_model, verbose=0, monitor='val_loss', save_best_only=True, mode='auto')

callbacks = [checkpoint, reducelrp]
```

**Definimos la red neuronal**

```python
mlp = Sequential()
mlp.add(Dense(16,
             input_dim=13,
             kernel_initializer='uniform',
             kernel_regularizer=regularizers.l2(0.001),
             activation='relu'))
mlp.add(Dropout(0.25))
mlp.add(Dense(8,
             kernel_initializer='uniform',
             kernel_regularizer=regularizers.l2(0.001),
             activation='relu'))
mlp.add(Dropout(0.15))
mlp.add(Dense(1,
             kernel_initializer='uniform',
             activation='sigmoid'))
mlp.compile(loss='binary_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
mlp.summary()
```

**Entrenando la red**  
Luego de varias pruebas se observo que la función de activación *tanh* funcionó de mejor manera para este problema.

```python
%%time
print("Tiempo de entrenamiento...!")
history = mlp.fit(X_train, 
                  y_train,
                  validation_split=0.33,
                  epochs=50,
                  batch_size=7,
                  callbacks=callbacks,
                  verbose=0)
plt.plot(history.history['accuracy'], label='train accuracy')
plt.plot(history.history['loss'], label='train loss')
plt.plot(history.history['val_accuracy'], label='valid accuracy')
plt.plot(history.history['val_loss'], label='valid loss')
plt.legend()
plt.show()
print("Entrenamiento finalizado")
```

```python
print("Evaluación de modelo con datos de entrenamiento")
score = mlp.evaluate(X_train, y_train, verbose=0)
print("score %s: %.2f%%" % (mlp.metrics_names[1], score[1]*100))

print("Evaluación de modelo con datos de prueba")
score = mlp.evaluate(X_test, y_test, verbose=0)
print("score %s: %.2f%%" % (mlp.metrics_names[1], score[1]*100))
```

**Carga y sumarización del mejor modelo entrenado**

```python
# Cargamos el modelo a partir del archivo generado y mostramos un resumen del mismo
final_mlp = tf.keras.models.load_model("mlp_hearthdisease.h5")
final_mlp.build()
final_mlp.summary()
```

**Predicciones utilizando el modelo**  
A partir del mejor modelo almacenado como parte de la funcionalidad del checkpoint se utilizara para realizar predicciones en el conjunto de datos de pruebas.

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print("Prueba de predicción en datos no observados")
# Predicción aproximada
y_pred_class = final_mlp.predict_classes(X_test)
# Probabilidad
y_pred_prob = final_mlp.predict(X_test)

for idx, label in enumerate(y_test):
    print ("%s    |   %s  | %s |   %.2f%%" % (str(idx), str(label), str(y_pred_class[idx]), float(y_pred_prob[idx])*100))

print ("Exactitud del modelo: %.2f%%" %  (accuracy_score(y_test, y_pred_class)*100))
```
