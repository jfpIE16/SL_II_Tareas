***Throughput***, es el volumen de información neto que fluye a traves de un sistema; para nuestro caso específico es la cantidad de información consumida por los usuarios en la red 3G y LTE.  
Se puede calcular de la siguiente manera: $$ T_{total} = \frac{TX_b + RX_b}{interval(tiempo)} $$
Interpretado como la suma total de bytes dividido el intervalo de tiempo definido, nuestro dataset contiene dicho valor en la unidad $Mbps$ mega bits por segundo.

**Propuesta**, el conjunto de datos utilizados contiene la información de un mes *x* con una granularidad de 15 mins para cada observación se tiene el *throughput* total de la red 3G y LTE en ese tiempo. Se propone construir una red neuronal recurrente del tipo *LSTM* (Long short-term memory) para que dada una secuencia definida del *throughput* en un período de tiempo se pueda predecir la secuencia siguiente. Se puede aplicar para detectar anomalías en la red de tráfico.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')

import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error

from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.callbacks import ModelCheckpoint, EarlyStopping
import tensorflow as tf
import keras
```

## Exploración del conjunto de datos

```python
tp_data = pd.read_csv('TP-05CENGT.csv', names=['Timestamp', 'Throughput(Mbps)'])
tp_data.head()
```

```python
tp_data.describe()
```

```python
plt.figure(figsize=(15,10))
plt.plot(tp_data['Throughput(Mbps)'].iloc[75:125], label='Throughput mes x')
plt.xlabel('Períodos de tiempo 15 min')
plt.ylabel('Throughput en Mbps')
plt.legend()
plt.show()
```

**Convertimos conjunto de datos a estructura de numpy**

```python
data = tp_data["Throughput(Mbps)"].values.astype(np.float32)
data = np.reshape(data, (-1,1))
data[0:6]
```

**Escalamiento de los datos y separación en datos de prueba y entrenamiento**  
Se aplicará una escala entre -1 y 1 para los datos, también se convertirán los datos a secuencias para poder trabajar de buena forma con el modelo RNN propuesto.

```python
scaler = MinMaxScaler(feature_range=(-1, 1))
dataset = scaler.fit_transform(data)
dataset[0:6]
```

```python
train_size = int(len(dataset) * 0.8)
test_size = len(dataset) - train_size
train, test = dataset[0:train_size, :], dataset[train_size:len(dataset),:]
print(f"Observaciones de entrenamiento: {len(train)}")
print(f"Observaciones de prueba: {len(test)}")
```

**Conversión a secuencias**  
Dado que la data se maneja con granularidad de 15$min$ se tomaran secuencias con tamaño 12, para obtener una secuencia de 3$horas$.

```python
def dataset2sequence(data, sequence=12):
    x, y = [], []
    for i in range(len(data)-sequence):
        a = dataset[i:(i+sequence), 0]
        x.append(a)
        y.append(dataset[i + sequence, 0])
    return np.array(x), np.array(y)
```

```python
trainx, trainy = dataset2sequence(train)
testx, testy = dataset2sequence(test)

trainy = trainy.reshape(-1, 1)
testy = testy.reshape(-1, 1)
```

```python
trainx.shape, trainy.shape
```

La red LSTM recibe los datos de entrada con una estructura de la forma:  
$[observaciones, granularidad, características]$

```python
train_x = trainx.reshape((trainx.shape[0], trainx.shape[1], 1))
test_x = testx.reshape((testx.shape[0], testx.shape[1], 1))

train_x.shape, test_x.shape
```

## Definición RNN (LSTM)

```python
#Early stopping
earlystop_monitor = EarlyStopping(monitor='val_loss', verbose=1, patience=30, baseline=0.4,)
#Model checkpoint
best_model = 'lstm_throughput.h5'
checkpoint = ModelCheckpoint(best_model, verbose=0, monitor='val_loss', save_best_only=True, mode='auto')
callbacks = [checkpoint]
```

```python
lstm = Sequential()
lstm.add(LSTM(units = 12, input_shape=(12,1),
             return_sequences=False))
lstm.add(Dense(1))
lstm.compile(loss='mae',
            optimizer='adam',
            metrics=['mse'])
lstm.summary()
```

**Entrenando la red**

```python
%%time
print("Tiempo de entrenamiento...!")
history = lstm.fit(train_x,
                   trainy,
                   epochs=50,
                   batch_size=1,
                   validation_data=(test_x, testy),
                   callbacks=callbacks,
                   verbose=1
                )
print("Entrenamiento finalizado")
```

```python
plt.plot(history.history['loss'], label='train loss')
plt.plot(history.history['val_loss'], label='valid loss')
plt.legend()
plt.show()
```

**Carga del mejor modelo entrenado para realizar predicciones y validaciones**

```python
# Cargamos el modelo a partir del archivo generado y mostramos un resumen del mismo
final_lstm = tf.keras.models.load_model("lstm_throughput.h5")
final_lstm.build()
final_lstm.summary()
```

```python
# Predicciones de conjuntos de datos
trainPredict = final_lstm.predict(train_x)
testPredict = final_lstm.predict(test_x)

# Invertir las predicciones con el scaler
trainPredict = scaler.inverse_transform(trainPredict)
trainY = scaler.inverse_transform(trainy)
testPredict = scaler.inverse_transform(testPredict)
testY = scaler.inverse_transform(testy)

# Calcular el RMSE
trainScore = np.sqrt(mean_squared_error(trainy[:, 0], trainPredict[:,0]))
print('Train Score: %.2f RMSE' % (trainScore))
testScore = np.sqrt(mean_squared_error(testy[:, 0], testPredict[:,0]))
print('Test Score: %.2f RMSE' % (testScore))
```
